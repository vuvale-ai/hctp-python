## 3 — The Math

Inspect the helix geometry at any σ value.

```python
import math

print("Helix radius R(σ) at key milestones:")
for s in [0.0, 0.25, 0.5, 0.75, 1.0]:
    R = helix_radius(s)
    pt = ideal_point(s)
    print(f"  σ={s:.2f}  R={R:.4f}  point=[{pt[0]:+.4f}, {pt[1]:+.4f}, {pt[2]:.4f}]")

print()
print("Ren K-vector stats at graduation:")
K = ren_K_final
sigma = progress(K)
d = distance(K, sigma)
print(f"  K       = {K}")
print(f"  sigma   = {sigma:.4f}")
print(f"  dist    = {d:.4f}  (0 = perfect helix tracking)")
print(f"  focus   = {determine_focus(K)} — {CHECKPOINTS[determine_focus(K)]["name"]}  (lowest k)")
```

## 4 — Live Session (no LLM required for the framework)

Simulate a training session using mock LLM responses.

```python
# Mock LLM for demo purposes
def mock_llm(prompt):
    """Returns a quality Karpathy-style response for demo scoring."""
    return (
        "The error in my original solution was treating the closure variable as immutable. "
        "The fix is to use nonlocal to properly reference the enclosing scope. "
        "Refactored to production-grade Python with type hints and docstrings. "
        "The lesson: closures capture by reference, not by value. "
        "Self-propose: How does this connect to decorator factories at checkpoint B? "
        ""
    )

# Create a fresh learner
ren = LearnerSession("Ren_demo")
print(f"Start: {ren}")

# Run 3 sessions
for session_num in range(3):
    session_data = ren.start_session()
    print(f"
--- Session {session_num + 1} ---")
    print(f"Focus: {session_data["focus"]} | Breadcrumbs: {session_data["n_breadcrumbs"]}")

    for bc_prompt in session_data["breadcrumb_prompts"]:
        bc_response = mock_llm(bc_prompt)
        kl_prompt   = ren.karpathy_prompt_for(bc_prompt, bc_response)
        kl_response = mock_llm(kl_prompt)
        delta = ren.submit_karpathy(kl_response)
        print(f"  Δk = {[round(d,4) for d in delta]}")

    result = ren.finish_session()
    print(result)

print(f"
Final: {ren}")
print(f"Badge earned: {ren.badge}")
```

```python
# Plot the demo session
plot_helix(
    sigma_history=ren.sigma_history,
    label="Ren_demo",
    title="HCTP — Live Demo Session (3 sessions)",
    show=True,
)
```

## 5 — Persistence

Save and restore learner state between sessions.

```python
import json, tempfile, pathlib

with tempfile.NamedTemporaryFile(suffix=".json", delete=False) as f:
    path = f.name

ren.save(path)
print("Saved state:")
print(json.dumps(json.loads(pathlib.Path(path).read_text()), indent=2))

ren2 = LearnerSession.load(path)
print(f"
Restored: {ren2}")
print(f"Sessions completed: {ren2.sessions_completed}")
```
